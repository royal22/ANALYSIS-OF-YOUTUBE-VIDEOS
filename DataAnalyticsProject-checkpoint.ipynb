```python
# Importing all the modules required 
# -pandas used for data manipulation and analysis 
# -json uses human readable text to transmit data objects consisting of attribute-value pairs
# -textblob is used for Sentiment analysis here 
# -numpy is used for array manipulation
# -matplotlib is used for plotting
# -seaborn is used for data visualization
 
import pandas as pd
import json
from textblob import TextBlob
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
# Here we are reading the JSON and merging all the videos with the same category to aid in our category-wise analysis
# DataFrame is a 2 dimensional array structure
df = open("GB_category_id.json")
youtube_df = df.read()
df.close()
df1 = json.loads(youtube_df)
cgs = pd.DataFrame([
    {'category_id':x['id'],
     'category':x['snippet']['title'],
     'assignable':x['snippet']['assignable']
    } for x in df1['items']
])
cgs
```

```python
# Here we are changing all the categories into integer type
# We are casting the entire pandas object to the same type
cgs.head()
cgs['category_id'] = cgs['category_id'].astype(int)
# Reading the csv files
GB_videos = pd.read_csv('GBvideos.csv',error_bad_lines=False)
comments_GB = pd.read_csv('GBcomments.csv',error_bad_lines=False)
GB_videos['category_id'] = GB_videos['category_id'].astype(int)
```

```python
# Displaying the top 5 rows of GB_videos
GB_videos.head()
```

```python
# Getting the data and formating it to the format "%d.%m"
GB_videos['date'] =  pd.to_datetime(GB_videos.date, format="%d.%m")
# Merging all the rows with the same category_id
GB_videos=GB_videos.merge(cgs, on='category_id')
```

```python
# We are using the textblob library to calculate sentiments
# The sentiment property returns a namedtuple of the form Sentiment(polarity, subjectivity).
# Here we are using the polarity
def sc(txt):
    try:
        return TextBlob(txt).sentiment.polarity
    except:
        return None
```

```python
# Applying the sentiment calculator function
comments_GB['polarity']=comments_GB['comment_text'].apply(sc)
```

```python
# Classifying the videos based on polarity
def polarity_sent(p):
    if p <=-.5:
        return "Very negative"
    if p >-.5 and p<0:
        return "Negative"
    if p == 0:
        return "Neutral"
    if p > 0 and p<0.5:
        return "Positive"
    if p >=0.5:
        return "Very positive"

def grouped_weighted_avg(values, weights, by):
   return (values * weights).groupby(by).sum() / weights.groupby(by).sum()
```

```python
# We're grouping the comments by video_id and calculating the mean to get the video sentiment
sentiment_of_video = comments_GB.groupby("video_id").mean()
sentiment_of_video['video_id'] = sentiment_of_video.index
sentiment_of_video.head()
```

```python
# We are merging the rows on video_id
GB_videos=GB_videos.merge(sentiment_of_video, on="video_id")
# We are adding the polarity into the data frame as video sentiment
GB_videos['sentiment_of_video']=GB_videos['polarity'].apply(polarity_sent)
# We are creating a csv file where we're storing the aggregated sentiment
sentiment_of_video['sentiment']=sentiment_of_video['polarity'].apply(polarity_sent)
sentiment_of_video.to_csv('video_comments_aggregated_sentiment.csv')
```

```python
print("Analysing average & total likes per category\n")
# We are finding the average likes per category and plotting it for easy visualisation.
# From the graph, we can see which category is most liked on the basis of likes
# Note : average is a better metric of comparison as more videos in a category lead to a higher number of view

GB_vid_last = GB_videos.sort_values('likes_x', ascending=False).drop_duplicates('video_id').sort_index()
GB_vid_last.groupby('category').likes_x.mean().plot(kind="bar")
print(GB_vid_last.groupby('category').likes_x.mean())
plt.title("Average Likes per Category")
plt.show()
print("Music Catergory seeems to have the most likes on average.")
```

```python
# We are finding the total likes per category and plotting the same for easy visualization
GB_vid_last.groupby('category').likes_x.sum().plot(kind="bar")
print(GB_vid_last.groupby('category').likes_x.sum())
plt.title("Total Likes per Category")
plt.show()
print("Music seems to be liked the most in case of average likes and total likes. Maybe because Music is one of the quickest of form  of entertainment, it's only minutes long.")
```

```python
# We are finding the average number of comments per category and plotting the same for easy visualization
print("Analysing average & total number of comments per category\n")
GB_vid_last.groupby('category').comment_total.mean().plot(kind="bar")
print(GB_vid_last.groupby('category').comment_total.mean())
plt.title("Average Number of Comments per Category")
plt.show()
```

```python
# We are finding the total dislikes per category and plotting the same for easy visualization
GB_vid_last.groupby('category').dislikes.sum().plot(kind="bar")
print(GB_vid_last.groupby('category').dislikes.sum())
plt.title("Total Dislikes per Category")
plt.show()
print("We can see that Music, Comedy and Entertainment are the catergories with the top most dislikes in case of both averge dislikes and total dislikes. We notice that these catergories also belong to some of the most liked catergories, indicating that they are polarized. This shows the taste of people in these categories is diverse.")
```
